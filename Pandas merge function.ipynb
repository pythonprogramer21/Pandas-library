```python
df=pd.DataFrame({'ID':[1,2,3,],'Class':[9,10,11,12]})
df
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D']})
df1
```

```python
pd.merge(df1,df,on='ID')
```

```python
df2=pd.DataFrame({'ID':[1,2,3,6],'Class':[9,10,11,12]})
df2
```

```python
pd.merge(df2,df1,on='ID',how='outer')
```

```python
pd.merge(df2,df1,on='ID',how='outer',indicator='True')
```
